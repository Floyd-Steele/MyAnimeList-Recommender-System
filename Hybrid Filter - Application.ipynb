```python
import pandas as pd
from fuzzywuzzy import process
```

```python
#Import anime lookup and cosine similarity tables 
```

```python
anime = pd.read_csv("anime_lookup_table.csv",index_col = 0)
cosine = pd.read_csv("cosine_table.csv",index_col = 0)
```

```python
#Build recommender function with first row displaying data for anime input - select metadata to display
```

```python
anime.columns
```

```python
def Recommender(name,cosine=cosine,anime=anime):
    selected_item = process.extractOne(name,anime["title_english"])
    indices = list(cosine.loc[selected_item[2]].sort_values(ascending = False).head(21).index)
    results = anime.loc[[int(i) for i in indices],:]
    return results[['title','title_english','type','episodes','status','rating','aired_from_year','score','scored_by']]   
```
